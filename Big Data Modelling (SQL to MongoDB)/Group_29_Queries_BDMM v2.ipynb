- Which two items get ordered together the most?

__Steps:__  
- Perform a self - join in invoice lines collection
- Project necessary data (InvoiceID, StocktemID1 StockItemID2, InvoiceLineID)
- Unwind the merged invoiceline to get the complete pais data (Stock1 - Stock2 - InvoiceId) - otherwise we might have (StockID1, {StockID2, StockID3 ...} , InvoiceID)
- Project necessary data (InvoiceID, StocktemID1 StockItemID2, InvoiceLineID)
- Filter to remove StockID1 == StockID2, otherwise they would always be in the same order and the results would be biased/wrong  
- Group by StockID1 and StockID2 to count the most common combination of products
- Sort the results and limit by one since we only want the most sold pair (2 items bought in the same InvoiceID)

__Note:__
- This query takes a lot of times and needs to be further optimized (timeit was not used because it was overheating the computer when running the query)

```python
query_0 = {
        "$lookup":{
           "from": "invoicelines",
           "localField": "InvoiceID",
           "foreignField": "InvoiceID",
           "as": "invoice_info"
        }
    }
query_1 = {
    "$project": {
        '_id': 1,
         'InvoiceID': 1,
         'StockItemID': 1,
        'invoice_info.StockItemID':1,
        'invoice_info.InvoiceID':1,'InvoiceLineID':1
    }}



query_2 = {'$unwind':'$invoice_info'}


query_3 = {'$project': 
           {'_id':0,'ProductID':'$StockItemID', 'Prod_2':'$invoice_info.StockItemID','InvoiceLineID':'$InvoiceLineID', \
           'InvoiceID':'$InvoiceID'}}
      
query_4 = {'$match':{'$expr':{'$ne':["$ProductID", "$Prod_2"]}}}

query_5 = {'$group': {'_id': {'ProductID':'$ProductID', 'Prod_2':'$Prod_2'} ,'count': { '$sum': 1}}}
        
query_6 = {'$sort':{'count':-1}}

limit = {'$limit':1}

pipeline = [query_0,query_1,query_2,query_3,query_4,query_5,query_6,limit]
r = db.invoiceslines_01.aggregate(pipeline)

pprint(list(r))
```

- For each customer category which 3 items have the ordered the most?

Steps: 
- Get a list of all customer categories
- For each category in CustomerCategory: 
    - Filter collection by category
    - Unwind InvoiceInfo
    - Project necessary data (CustomerCategory, ProductId, Quantity)
    - Groupby CustomerCategory and ProductID and sum the total quantities bought
    - Sort results (descending) and limit by 3 since we want the most bought
    - Store results of specifc category in a list 
- Return answer with all the categories

- Without index

```python
%%timeit
categories = []
for i in list(db.customercategories.find()):
    categories.append(i['CustomerCategoryName'])
    
    
answer = []
for i in categories:
    query_00 = {'$match': {'CustomerCategory':i}}
    query_0 = {'$unwind':'$invoice_info'}
    query_1 = {'$project': 
               {'_id':0,'ProductID':'$invoice_info.StockItemID','CustomerCategory':'$CustomerCategory', \
                'Quantity':'$invoice_info.Quantity'}}

    query_2 = {'$group': 
               {'_id':{'Category':'$CustomerCategory','ProductID':'$ProductID'}, \
                'Total Orders': {'$sum':'$Quantity'}}}

    query_3 = {'$sort': {'Total Orders':-1, 'Category':1}}

    query_4 = {'$project': {'_id':False, 'ProductID':'$_id.ProductID','Category':'$_id.Category', \
                            'Total Orders':'$Total Orders'}}

    limit = {'$limit':3}

    pipeline = [query_00, query_0,query_1,query_2,query_3,query_4,limit]
    r = list(db.invoices_embed.aggregate(pipeline))
    answer.append(r)
pprint(answer)
```

What is the current stock of each stockgroup?

Steps:
- Get stockgroup information into stockitemstransactions  
- Project necessary data (StockItemID, Quantity,StockzGroup InvoiceLineID)  
- Unwind information
- Group by stockgroup and sum quantitites (It's assumed that the sum of the quantities gives the current stock since negative flows of goods represent customer sales and positive flows represent suppliers purchases,  eg: if the commpany bought 100 units of Product 1 and sold, 20, 10, 5 to customers A,B and C then the final stock would be (100 - (20+10+5)) = 65
- Return relevant information

```python
%%timeit
query_0 = {
        "$lookup":{
           "from": "stockgroup_aux",
           "localField": "StockItemID",
           "foreignField": "StockItemID",
           "as": "stockgroup"
        }
    }

query_1 = {
    "$project": {
        '_id': 1,
         'StockItemID': 1,
         'Quantity': 1,
        'StockGroup':'$stockgroup.StockGroup','InvoiceLineID':1
    }}

query_2 = {'$unwind':'$StockGroup'}


query_3 = {'$group':{'_id':{'StockGroup':'$StockGroup'},'CurrentStock': {'$sum':'$Quantity'}}}

query_4 = {'$project': {'_id':False,'Stockgroup':'$_id.StockGroup',
                        'Current Stock':'$CurrentStock'}}



pipeline = [query_0,query_1,query_2,query_3,query_4]
r = db.new_stockitemstransactions_01.aggregate(pipeline)

pprint(list(r))
```

5.A What is the name of the sales person with the largest sum of invoice values in 2013 (person whose customers paid the most money)?

Firstly, since the amount spent by the customers was not part of the collection invoices_embed, we performed the embedding of this field into it. We chose this approach instead of referencing since it facilitates querying. Also, we are only embedding a single field so it will not overload the collection.

```python
%%timeit
#Embed TransactionAmount into invoices for querying
query_22 = {
        "$lookup":{
           "from": "customertransactions_03",
           "localField": "InvoiceID",
           "foreignField": "InvoiceID",
           "as": "transactions"
        }
    }

query_23 = {
    "$unwind": "$transactions"
}

query_e = {
    '$project': {
        "transactions.TransactionAmount" : 1
    }
}

query_24 = { 
        "$merge" : {
            "into" : "invoices_embed"
        } 
    }

pipeline = [query_22, query_23, query_e ,query_24]

f =list(db.invoices_embed.aggregate(pipeline))
```

We assumed that the sum of invoices corresponds to the sum of customer transactions

```python
%%timeit
#WithoutIndex
query_2 = {'$match' : { "$expr": { "$eq": [{ "$year": "$InvoiceDate" }, 2013] }}} #Get only invoices regarding 2013

query_3 = {
    '$unwind': '$transactions'
}


query_4 = {
    '$project': {
        'SalespersonPersonName' : 1,
        "Invoices": "$transactions.TransactionAmount"  #Get the value of the transactions
}}

query_5 = {
    "$group": {
        "_id": "$SalespersonPersonName",
        "Sum_Invoices": {"$sum": "$Invoices"}  #Sum the transactions and group by Sales person name
    }
}


query_6 = {
    "$sort": {"Sum_Invoices": -1}  #Sort descending
}
query_7 = {
    '$limit': 1 #Show only the sales person with the highest value
}


pipeline = [query_2,query_3,query_4 ,query_5, query_6, query_7]

r = db.invoices_embed.aggregate(pipeline)

result = list(r)

pprint(result)
```

```python
%%timeit
#WithIndex - Slower
query_2 = {'$match' : { "$expr": { "$eq": [{ "$year": "$InvoiceDate" }, 2013] }}}  #Get only invoices regarding 2013

query_3 = {
    '$unwind': '$transactions'
}


query_4 = {
    '$project': {
        'SalespersonPersonName' : 1,
        "Invoices": "$transactions.TransactionAmount"  #Get the value of the transactions
}}

query_5 = {
    "$group": {
        "_id": "$SalespersonPersonName",
        "Sum_Invoices": {"$sum": "$Invoices"} #Sum the transactions and group by Sales person name
    }
}


query_6 = {
    "$sort": {"Sum_Invoices": -1}  #Sort descending
}
query_7 = {
    '$limit': 1 #Show only the sales person with the highest value
}


pipeline = [query_2,query_3,query_4 ,query_5, query_6, query_7]

r = db.invoices_embed.aggregate(pipeline)

result = list(r)

pprint(result)
```
